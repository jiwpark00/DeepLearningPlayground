First Steps using Keras 

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd

seed = 123
np.random.seed(seed)
```

```python
dataset = pd.read_csv(r'https://raw.githubusercontent.com/a-coders-guide-to-ai/a-coders-guide-to-neural-networks/master/data/diabetes.csv')
```

```python
dataset.shape
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
dataset.columns
```

```python
X = dataset[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]

Y = dataset[['Outcome']]
```

```python
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print(model.metrics_names[1], scores[1]*100)
```

Improving accuracy

```python
model2 = Sequential()
model2.add(Dense(100, input_dim=8, activation='relu'))
model2.add(Dense(20, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))

model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model2.fit(X, Y, epochs=150, batch_size=10, verbose=1)
```

```python
scores2 = model2.evaluate(X, Y)
print(model2.metrics_names[1], scores2[1]*100)
```

```python
model3 = Sequential()
model3.add(Dense(100, input_dim=8, activation='tanh'))
model3.add(Dense(20, activation='relu'))
model3.add(Dense(1, activation='sigmoid'))

model3.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])

model3.fit(X, Y, epochs=150, batch_size=10, verbose=1)
```

```python
scores3 = model3.evaluate(X, Y)
print(model3.metrics_names[1], scores3[1]*100)
```

```python
model4 = Sequential()
model4.add(Dense(1000, input_dim=8, activation='relu'))
model4.add(Dense(200, activation='relu'))
model4.add(Dense(1, activation='sigmoid'))

model4.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])

model4.fit(X, Y, epochs=150, batch_size=10, verbose=1)
```
